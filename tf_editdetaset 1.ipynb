```
import numpy as np
import os
import os.path
import glob
import shutil
```

```
dir_path = "/content/drive/MyDrive/後期/きゅうり10/きゅうり切り取り400-TFRecords-export"

os.mkdir(os.path.join(dir_path, 'train'))
os.mkdir(os.path.join(dir_path, 'val'))

trainRatio = 0.9
```

```
files = glob.glob(os.path.join(dir_path, '*.tfrecord'))
files = np.array(files)

numfile = files.shape[0]
randindex = np.random.permutation(numfile)

numtrain = np.floor(numfile*trainRatio)
cnt = 1

for index in randindex:
  file = files[index]
  
  if cnt <= numtrain:
    shutil.move(file, os.path.join(dir_path, 'train/'))
  else:
    shutil.move(file, os.path.join(dir_path, 'val/'))
  cnt += 1
```

```
trainfiles = glob.glob(os.path.join(dir_path, 'train/*.tfrecord'))
valfiles = glob.glob(os.path.join(dir_path, 'val/*'))

for i,f in enumerate(trainfiles):
  os.rename(f, os.path.join(dir_path, 'train/train_' + '{0:04d}'.format(i) + '.tfrecord'))
  
for i,f in enumerate(valfiles):
  os.rename(f, os.path.join(dir_path, 'val/val_' + '{0:04d}'.format(i) + '.tfrecord'))
```
